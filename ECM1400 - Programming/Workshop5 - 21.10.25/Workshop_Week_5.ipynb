Create a simple program to average the numbers input by a user. (recap from workshop week 3 if you need it).

```python
def Average(nums):
    total = 0
    for num in nums:
        total += num
    return total / len(nums)

userInput = input("Enter some numbers separated by commas: ")
try:
    userInputSplit = userInput.split(",")

    nums = []
    for num in userInputSplit:
        nums.append(float(num))
    
    average = Average(nums)
    print(f"The average of your nums is {average}") 
except:
    print("Enter the numbers correctly!!")
```
